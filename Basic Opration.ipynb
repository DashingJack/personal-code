### Set Operations

```python
set1 = {1,3,5,7,1,9}
print(set1)
```

```python
set1.add(88)
print(set1)
```

```python
set2 = {2,4,6,9,11}
set3 = set1.union(set2)
set3
```

```python
set3 = set1.intersection(set2)
set3
```

```python
set3 = set1.difference(set2)
print(set3)
```

```python
set3 = set1.isdisjoint(set2)
print(set3)
```

```python
x = {"a", "b", "c"}
y = {"f", "e", "d", "c", "b", "a"}
z = x.issubset(y)
print(z)
```
