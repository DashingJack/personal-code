```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
print(df.isnull().sum())
```

```python
df = df.drop("Cabin", axis=1 )
```

```python
df.head()
```

```python
print(df.isnull().sum())
```

```python
df["Age"] = df["Age"].replace(np.NaN, df["Age"].mean())
print(df.isnull().sum())
```

```python
df["Age"] = df["Age"].replace(np.NaN, df["Age"].median())
print(df.isnull().sum())
```

```python
import statistics
df["Age"] = df["Age"].replace(np.NaN, statistics.mode(df["Age"]))
print(df.isnull().sum())
```

```python
df.dropna(inplace=True)
print(df.isnull().sum())
```

```python
df1 = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
```

```python
# Fitting the data to the imputer object
df1["Age"] = imputer.fit_transform(df1[['Age']])
```

```python
imputer2 = SimpleImputer(missing_values = np.nan, strategy = 'median')
```

```python
# Fitting the data to the imputer2 object
df1["Age"] = imputer2.fit_transform(df1[['Age']])
```

```python
imputer3 = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
```

```python
# Fitting the data to the imputer object
df1["Age"] = imputer3.fit_transform(df1[['Age']])
```

```python
imputer4 = SimpleImputer(missing_values = np.nan, strategy = 'constant', fill_value = 'D')
```

```python
# Fitting the data to the imputer object
df1["Cabin"] = imputer4.fit_transform(df1[['Cabin']])
```

```python
print(df1.isnull().sum())
```
