```python
n=3
for i in range(n):
  for j in range(i+1):
    print("*",end=" ")
  print("\n")
```

```python
for i in range(n):
  for j in range(n-i):
    print("*",end=" ")
  print("\n")
```

```python
m = 2*n-1
for i in range(n):
  for j in range(0,i):
    print(" ",end="")
  for j in range(i, m-i):
    print("*",end=" ")
  print("\n")
for i in range(1, n):
  for j in range(0,n-i-1):
    print(" ",end="")
  for j in range(n-i,m+i-1):
    print("*",end=" ")
  print("\n")
```
