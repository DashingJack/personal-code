#Write a program to check whether the given number is prime or not.

```python
num = 7

flag = False
if num>1:
  for i in range(2, num):
    if(num%i == 0):
      flag = True
      break
if flag:
  print("Number is not prime")
else:
  print("Number is prime")
```

#Write a program to print Fibonacci series up to number given by user.

```python
num = 17

f = [0, 1]
if num>1:
  for i in range(2, num+1):
    f.append(f[i-1] + f[i-2])
print(f)
```

#Write a python program to find the factorial of a given number using recursion.

```python
num = 5
def fact(n):
  if n == 0:
    return 1
  return n * fact(n-1)
fact(num)
```

#Write a program to check whether the given number is Armstrong or not.

```python
num = 1772

tnum = num
dgt = 0
while(tnum>0):
  tnum //= 10
  dgt += 1

tnum = num
temp = 0
for i in range(dgt):
  temp += (tnum%(10))**dgt
  tnum //= 10

if (temp==num):
  print("Given number is armstrong number")
else:
  print("Given number is not armstrong number")
```

#Write a program to check whether the given number is palindrome or not.

```python
num = 1771
reverse = 0
temp = num
while (temp != 0):
    reverse = (reverse * 10) + (temp % 10)
    temp = temp // 10

if(reverse == num):
  print("Number is palindrom")
else:
  print("Number is not palindrom")
```
